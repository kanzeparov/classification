```python
from pydub import AudioSegment
import warnings
import  keras
from keras.utils import *
from  keras.layers import Activation, Dense, Dropout, Conv2D, Flatten, MaxPooling2D
from keras.models import Sequential
import librosa
import librosa.display
import numpy as np
import pandas as pd
import random
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('meta\meta.txt', header = None, sep = '\t')
data.head(5)
```

```python
data.shape
```

Find all types

```python
s = set(data[4])
dics = dict()
for i in range(len(s)):
    dics[s.pop()] = i
print(dics)
```

```python
val = data
val["path"] = 'audio\\' + data[0]
```

Merge sounds if they less than 20 seconds

```python
def combine_sound(path):
    s = "temp.wav"
    sound1 = AudioSegment.from_wav(path)
    sound2 = AudioSegment.from_wav(path)
    combined_sounds = sound1 + sound2
    combined_sounds.export(s, format="wav")
    print(combined_sounds.duration_seconds)
    if(combined_sounds.duration_seconds<12):
        combine_sound(s)
    return s
```

```python

D  =  []  # Dataset# Dataset

for row in val.itertuples():
    
        print(row.path)
        y, sr = librosa.load(combine_sound(row.path),duration=11.87)  
        ps = librosa.feature.melspectrogram(y=y, sr=sr)
        print(ps.shape)
        print(row[5])
        D.append( (ps, dics[row[5]]) )
```

```python
print("Number of samples: ", len(D))
```

```python
ps.shape
```

```python

dataset  = D
random.shuffle(dataset)

train  =  dataset[:10100]
test  =  dataset[10100:]

X_train, y_train  = zip(*train)
X_test, y_test  = zip(*test)

  # Reshape for CNN input# Reshape 
X_train = np.array([x.reshape( (128, 512, 1) ) for x in X_train])
X_test = np.array([x.reshape( (128, 512, 1) ) for x in X_test])

# One-Hot encoding for classes
y_train = np.array(keras.utils.to_categorical(y_train, 10))
y_test = np.array(keras.utils.to_categorical(y_test, 10))
```

```python
X_train.shape
```

```python

model  =  Sequential()
input_shape=(128, 512, 1)

model.add(Conv2D(24, (5, 5), strides=(1, 1), input_shape=input_shape))
model.add(MaxPooling2D((4, 2), strides=(4, 2)))
model.add(Activation('relu'))

model.add(Conv2D(48, (5, 5), padding="valid"))
model.add(MaxPooling2D((4, 2), strides=(4, 2)))
model.add(Activation('relu'))

model.add(Conv2D(48, (5, 5), padding="valid"))
model.add(Activation('relu'))

model.add(Flatten())
model.add(Dropout(rate=0.5))

model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(rate=0.5))

model.add(Dense(10))
model.add(Activation('softmax'))
```

```python

model.compile(optimizer="Adam",loss="categorical_crossentropy",metrics=['accuracy'])

model.fit(x=X_train, y=y_train,epochs=12,batch_size=128,validation_data= (X_test, y_test))
score = model.evaluate(x=X_test,y=y_test)

print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
